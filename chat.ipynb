```python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

# 1. Load the .env file
load_dotenv()

# 2. Get the key and setup the client
my_key = os.getenv("ANTHROPIC_API_KEY")
client = Anthropic(api_key=my_key)

# 3. Set the latest model name
model_name = "claude-3-5-sonnet-latest"

print("✅ Setup Complete!")   
```

```python
# Create an empty list to store the chat history
messages = []

print("--- Chatbot Started (Type 'quit' to stop) ---")

while True:
    user_input = input("You: ")
    
    if user_input.lower() in ["quit", "exit"]:
        print("Goodbye!")
        break

    # Add your message to the history
    messages.append({"role": "user", "content": user_input})

    # Send the history to Claude
    response = client.messages.create(
        model=model_name,
        max_tokens=1024,
        messages=messages
    )

    # Get Claude's text and add it to history
    ai_text = response.content[0].text
    messages.append({"role": "assistant", "content": ai_text})

    print(f"Claude: {ai_text}")
```
